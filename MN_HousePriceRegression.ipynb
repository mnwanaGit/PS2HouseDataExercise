```python
import pandas as pd #import statements
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
train = pd.read_csv('/train.csv') #loading in data
test = pd.read_csv('/test.csv')
```

```python
corrs = abs(train.corr()) #correlation logic: choose the variables with the 10 highest absolute value of correlation coefficient
print(corrs)
```

From correlation sheet, top 10 variables with highest absolute value of correlation coefficient:

    1. OverallQual
    2. GrLivArea
    3. GarageCars
    4. GarageArea
    5. TotalBsmtSF
    6. 1stFlrSF
    7. FullBath
    8. TotRmsAbvGrd
    9. YearBuilt
    10. YearRemodAdd 

```python
x = train[['OverallQual', 'GrLivArea', 'GarageCars', 'GarageArea',
                           'TotalBsmtSF', '1stFlrSF', 'FullBath', 'TotRmsAbvGrd',
                           'YearBuilt', 'YearRemodAdd']]

y = train['SalePrice']

x_train, x_valid, y_train, y_valid = train_test_split(x, y, test_size=0.33, random_state=42) #creating our training/validation data
```

```python
reg = LinearRegression().fit(x_train, y_train) #fitting the linear regression model
```

```python
print("Training R^2 is:", reg.score(x_valid, y_valid))
```

```python
y_pred = reg.predict(x_train) #getting prediction (Example)

y_pred[0]
```

```python
test_predictors = test[['OverallQual', 'GrLivArea', 'GarageCars', 'GarageArea',
                           'TotalBsmtSF', '1stFlrSF', 'FullBath', 'TotRmsAbvGrd',
                           'YearBuilt', 'YearRemodAdd']] #pulling back the predictors we used to train the model

pred_prices = reg.predict(test_predictors) #predicting the true test data's sale prices
test['SalePrice(pred)'] = pred_prices

test.head()
```

```python
ship_out = pd.DataFrame() #creating a dataframe with only ID's and predicted saleprices to ship out

ship_out['Id'] = test['Id']
ship_out['SalePrice(pred)'] = test['SalePrice(pred)']
ship_out.head()
```

```python
ship_out.to_csv('/predictions.csv')
```
